```python
# =========================================================================
# File: skin_disease_detection.py
# Author: Gemini
# Description: This script trains a deep learning model to classify skin diseases
#              using a provided image dataset. It leverages transfer learning
#              with a pre-trained MobileNetV2 model to achieve high accuracy.
#              The script includes steps for data loading, model building,
#              training, evaluation, and making predictions on new images.
# =========================================================================

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt
import os

# --- Step 1: Set up dataset paths and parameters ---
# IMPORTANT: Update this path to the location of your 'archivenew' folder.
DATASET_PATH = '/content/archivenew/archivenew'
TRAIN_DIR = os.path.join(DATASET_PATH, 'train')
VAL_DIR = os.path.join(DATASET_PATH, 'val')

# Define model parameters
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
NUM_EPOCHS = 90 # You can increase this for better accuracy
LEARNING_RATE = 0.0001

# --- Step 2: Data Loading and Augmentation ---
# We use ImageDataGenerator for data loading and real-time data augmentation.
# This helps prevent overfitting and improves model generalization.

# Create data generators for training and validation datasets
# Rescale pixel values to [0, 1] for the model
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255) # Only rescaling for validation data

print("Loading training data...")
train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical' # 'categorical' for multi-class classification
)

print("Loading validation data...")
val_generator = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

# Get the number of classes from the training generator
NUM_CLASSES = train_generator.num_classes
class_names = list(train_generator.class_indices.keys())
print(f"Detected {NUM_CLASSES} classes: {class_names}")

# --- Step 3: Build the Model using Transfer Learning ---
# We will use MobileNetV2, a pre-trained model on ImageNet.
# This allows the model to leverage features learned from a large dataset.

# Load the MobileNetV2 model with pre-trained weights, excluding the top layers.
# The `include_top=False` argument removes the final classification layer.
base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)
)

# Freeze the base model layers to prevent their weights from being updated during training.
# This is a common practice in transfer learning.
base_model.trainable = False

# Add new custom layers on top of the pre-trained model for our specific task.
# GlobalAveragePooling2D layer reduces the feature maps to a single vector.
x = base_model.output
x = GlobalAveragePooling2D()(x)
# Add a dense layer for classification with a softmax activation.
predictions = Dense(NUM_CLASSES, activation='softmax')(x)

# Combine the base model and the new layers to create the final model.
model = Model(inputs=base_model.input, outputs=predictions)

# --- Step 4: Compile the Model ---
# We use the Adam optimizer and 'categorical_crossentropy' loss for multi-class problems.
# The 'accuracy' metric will track the performance.
model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Print a summary of the model architecture
model.summary()

# --- Step 5: Train the Model ---
print("\nStarting model training...")
history = model.fit(
    train_generator,
    epochs=NUM_EPOCHS,
    validation_data=val_generator,
    verbose=1
)

# --- Step 6: Evaluate the Model ---
print("\nEvaluating model on validation data...")
loss, accuracy = model.evaluate(val_generator)
print(f"Validation Loss: {loss:.4f}")
print(f"Validation Accuracy: {accuracy:.4f}")

# Plot training & validation accuracy and loss
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# --- Step 7: Save the Trained Model ---
MODEL_SAVE_PATH = 'skin_disease_model.keras'
model.save(MODEL_SAVE_PATH)
print(f"\nModel saved successfully to {MODEL_SAVE_PATH}")

# =========================================================================
# --- Step 8: Create a Prediction Function ---
# This function loads the trained model and makes a prediction on a new image.
# =========================================================================

def predict_skin_disease(image_path, model_path=MODEL_SAVE_PATH):
    """
    Predicts the skin disease from a new image.

    Args:
        image_path (str): The path to the new image file.
        model_path (str): The path to the saved model file.

    Returns:
        str: The name of the predicted skin disease class.
    """
    try:
        # Load the saved model
        loaded_model = tf.keras.models.load_model(model_path)

        # Load and preprocess the new image
        img = tf.keras.preprocessing.image.load_img(
            image_path, target_size=IMAGE_SIZE
        )
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0) # Create a batch
        img_array /= 255.0 # Rescale the image

        # Make the prediction
        predictions = loaded_model.predict(img_array)
        predicted_class_index = np.argmax(predictions[0])
        confidence = np.max(predictions[0]) * 100

        # Get the predicted class name
        predicted_class_name = class_names[predicted_class_index]

        print(f"\nPrediction for image: {image_path}")
        print(f"Predicted class: {predicted_class_name}")
        print(f"Confidence: {confidence:.2f}%")
        return predicted_class_name

    except Exception as e:
        print(f"An error occurred during prediction: {e}")
        return None

# =========================================================================
# --- Step 9: Example Usage of the Prediction Function ---
# =========================================================================

# Create a sample image for demonstration purposes.
# IMPORTANT: Replace 'path/to/your/test_image.jpg' with a real image path from your dataset.
# The image should be located outside the 'train' and 'val' folders.
# You can also use one of the images from your validation folder.
sample_image_path = os.path.join(VAL_DIR, class_names[0], os.listdir(os.path.join(VAL_DIR, class_names[0]))[0])

# To test with a specific image, uncomment the line below and provide a valid path.
# sample_image_path = 'path/to/your/new_image.jpg'

if os.path.exists(sample_image_path):
    print("\n--- Running example prediction ---")
    predict_skin_disease(sample_image_path)
else:
    print(f"\nSample image not found at {sample_image_path}. Please provide a valid path to test the prediction function.")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import os

# --- Configuration (Must match the training setup) ---
MODEL_SAVE_PATH = 'skin_disease_model.keras'
IMAGE_SIZE = (224, 224)

# 🥇 THE CORRECTED CLASS NAMES 🥇
class_names = [
    'Acne', 'Actinic Keratosis', 'Basal Cell Carcinoma', 'Chickenpox', 
    'Dermato Fibroma', 'Dyshidrotic Eczema', 'Melanoma', 'Nail Fungus', 
    'Nevus', 'Normal Skin', 'Pigmented Benign Keratosis', 'Ringworm', 
    'Seborrheic Keratosis', 'Squamous Cell Carcinoma', 'Vascular Lesion'
]

# =========================================================================
# --- Prediction Function ---
# =========================================================================

def predict_skin_disease(image_path, model_path=MODEL_SAVE_PATH, classes=class_names):
    """
    Predicts the skin disease from a new image using the saved model.
    """
    if not os.path.exists(model_path):
        print(f"Error: Model file not found at {model_path}. "
              f"Please train the model first or check the path.")
        return None

    try:
        # Load the saved model
        # NOTE: This assumes the model was saved with the same TensorFlow version.
        loaded_model = tf.keras.models.load_model(model_path)

        # Load and preprocess the new image
        img = load_img(image_path, target_size=IMAGE_SIZE)
        img_array = img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0) # Add batch dimension
        img_array /= 255.0 # Rescale

        # Make the prediction
        predictions = loaded_model.predict(img_array)
        predicted_class_index = np.argmax(predictions[0])
        confidence = np.max(predictions[0]) * 100

        # Get the predicted class name using the correct list
        predicted_class_name = classes[predicted_class_index]

        print(f"\nPrediction for image: {image_path}")
        print("-" * 35)
        print(f"Predicted Class: {predicted_class_name}")
        print(f"Confidence: {confidence:.2f}%")
        return predicted_class_name

    except Exception as e:
        print(f"An error occurred during prediction: {e}")
        return None

# =========================================================================
# --- Example Usage ---
# =========================================================================

# VITAL: Replace 'path/to/your/test_image.jpg' with a real, valid path.
sample_image_path = 'archivenew/archivenew/train/Squamous Cell Carcinoma/ISIC_0026388_0_630.jpeg' 

print("\n--- Running example prediction with 15 classes ---")

if os.path.exists(sample_image_path) and sample_image_path != 'path/to/your/test_image.jpg':
    predict_skin_disease(sample_image_path)
else:
    print("--- ACTION REQUIRED ---")
    print(f"Sample image not found or path not updated. Please set 'sample_image_path' to a valid image file to test the model.")
```
